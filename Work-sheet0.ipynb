```python
def convert_units():
    """
    Converts between length, weight, and volume units.
    Handles invalid input using try-except.
    """

    print("Choose conversion type:")
    print("1. Length")
    print("2. Weight")
    print("3. Volume")

    choice = input("Enter choice (1/2/3): ")

    try:
        value = float(input("Enter value to convert: "))

        if choice == "1":
            print("1. Meters to Feet")
            print("2. Feet to Meters")
            opt = input("Choose option: ")

            if opt == "1":
                print("Result:", value * 3.28084, "ft")
            elif opt == "2":
                print("Result:", value / 3.28084, "m")

        elif choice == "2":
            print("1. Kilograms to Pounds")
            print("2. Pounds to Kilograms")
            opt = input("Choose option: ")

            if opt == "1":
                print("Result:", value * 2.20462, "lb")
            elif opt == "2":
                print("Result:", value / 2.20462, "kg")

        elif choice == "3":
            print("1. Liters to Gallons")
            print("2. Gallons to Liters")
            opt = input("Choose option: ")

            if opt == "1":
                print("Result:", value * 0.264172, "gallons")
            elif opt == "2":
                print("Result:", value / 0.264172, "liters")

        else:
            print("Invalid conversion type")

    except ValueError:
        print("Invalid numeric input")

convert_units()
```

```python
def find_sum(lst):
    """Return sum of list"""
    return sum(lst)

def find_average(lst):
    """Return average"""
    return sum(lst) / len(lst)

def find_max(lst):
    """Return maximum"""
    return max(lst)

def find_min(lst):
    """Return minimum"""
    return min(lst)

try:
    numbers = list(map(float, input("Enter numbers separated by space: ").split()))

    print("Choose operation:")
    print("1 Sum")
    print("2 Average")
    print("3 Maximum")
    print("4 Minimum")

    op = input("Enter choice: ")

    if op == "1":
        print("Sum =", find_sum(numbers))
    elif op == "2":
        print("Average =", find_average(numbers))
    elif op == "3":
        print("Max =", find_max(numbers))
    elif op == "4":
        print("Min =", find_min(numbers))
    else:
        print("Invalid operation")

except:
    print("Invalid input")
```

```python
def extract_every_other(lst):
    return lst[::2]

# Example:
print(extract_every_other([1, 2, 3, 4, 5, 6]))
```

```python
def get_sublist(lst, start, end):
    return lst[start:end+1]  # +1 because slicing excludes the end index

# Example:
print(get_sublist([1, 2, 3, 4, 5, 6], 2, 4))
```

```python
def reverse_list(lst):
    return lst[::-1]

# Example:
print(reverse_list([1, 2, 3, 4, 5]))
```

```python
def remove_first_last(lst):
    return lst[1:-1]

# Example:
print(remove_first_last([1, 2, 3, 4, 5]))
```

```python
def get_first_n(lst, n):
    return lst[:n]

# Example:
print(get_first_n([1, 2, 3, 4, 5], 3))
```

```python
def get_last_n(lst, n):
    return lst[-n:]

# Example:
print(get_last_n([1, 2, 3, 4, 5], 2))
```

```python
def reverse_skip(lst):
    return lst[-2::-2]  # Start from second-to-last, step backwards by 2

# Example:
print(reverse_skip([1, 2, 3, 4, 5, 6]))
```

```python
def flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result

# Example usage:
nested_list = [[1, 2], [3, 4], [5]]
print(flatten(nested_list))
```

```python
def access_nested_element(lst, indices):
    element = lst
    for index in indices:
        element = element[index]  # Drill down step by step
    return element

# Example usage:
lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]
print(access_nested_element(lst, indices))
```

```python
def sum_nested(lst):
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

# Example usage:
nested_list = [[1, 2], [3, [4, 5]], 6]
print(sum_nested(nested_list))
```

```python
def remove_element(lst, elem):
    result = []
    for item in lst:
        if isinstance(item, list):
            # Recursively process sublists
            cleaned = remove_element(item, elem)
            if cleaned:  # Only add non-empty sublists
                result.append(cleaned)
        else:
            if item != elem:  # Keep only items not equal to elem
                result.append(item)
    return result

# Example usage:
lst = [[1, 2], [3, 2], [4, 5]]
elem = 2
print(remove_element(lst, elem))
```

```python
def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)  # Recursively count inside sublists
        else:
            if item == elem:
                count += 1
    return count

# Example usage:
lst = [[1, 2], [2, 3], [2, 4]]
elem = 2
print(count_occurrences(lst, elem))  # Output: 3
```

```python
def deep_flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))  # Recursively flatten sublists
        else:
            result.append(item)
    return result

# Example usage:
nested_list = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print(deep_flatten(nested_list))
```

```python
def average_nested(lst):
    def helper(sublist):
        total = 0
        count = 0
        for item in sublist:
            if isinstance(item, list):
                t, c = helper(item)  # Recursively process sublists
                total += t
                count += c
            else:
                total += item
                count += 1
        return total, count

    total, count = helper(lst)
    return total / count if count != 0 else 0

# Example usage:
nested_list = [[1, 2], [3, 4], [5, 6]]
print(average_nested(nested_list))
```

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# 1 A * A^-1 = I
A_inv = np.linalg.inv(A)
I = np.dot(A,A_inv)

print("A * A^-1 =\n", I)

# 2 AB ≠ BA
AB = np.dot(A,B)
BA = np.dot(B,A)

print("\nAB =\n", AB)
print("\nBA =\n", BA)

# 3 (AB)^T = B^T A^T
left = AB.T
right = np.dot(B.T, A.T)

print("\n(AB)^T =\n", left)
print("\nB^T A^T =\n", right)
```

```python
A = np.array([
    [2,-3,1],
    [1,-1,2],
    [3,1,-1]
])

B = np.array([-1,-3,9])

# Using inverse
A_inv = np.linalg.inv(A)
X = np.dot(A_inv,B)

print("Solution using inverse:")
print("x, y, z =", X)

# Using solve
solution = np.linalg.solve(A,B)

print("\nSolution using numpy solve:")
print("x, y, z =", solution)
```
